```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv('anaemia.csv')
```

```python
df
```

```python
df=df.drop(['Number'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['Anaemic']=le.fit_transform(df['Anaemic'])
```

```python
df
```

```python
mapping={'M':0,'F':1,}
df['Sex']=df['Sex'].map(mapping)
```

```python
# print(df[:,1:5])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
df.iloc[:,1:5]=ss.fit_transform(df.iloc[:,1:5])
```

```python
print(df.iloc[:,0])
```

```python
from sklearn.impute import KNNImputer
```

```python
# df=np.array(df)
```

```python
imputer=KNNImputer(n_neighbors=5,weights='uniform')
df=imputer.fit_transform(df)
```

```python
df[:,0]
```

```python
result=0
for i in df[:,0]:
    if i>=0.5:
        df[result,0]=1.0
    else:
        df[result,0]=0.0
    result+=1
```

```python
df[:,0]
```

```python
df
```

```python
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
x=df[:,:4]
```

```python
y=df[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
param_grid={
    'n_estimators':[100,10,50],
    'learning_rate':[0.1,0.2,0.5],
    'max_depth':[3,5,7],
    'random_state':[42,420,388]
}
```

```python
grid_search = GridSearchCV(XGBClassifier(), param_grid=param_grid, 
                           scoring='accuracy', cv=3, verbose=1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
algh=XGBClassifier(learning_rate=0.1,max_depth=3,n_estimators=10,random_state=42)
```

```python
algh.fit(x_train,y_train)
```

```python
algh.score(x_test,y_test)
```

```python
param_grid={
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': [1, 0.1, 0.01]
}
```

```python
grid_search2 = GridSearchCV(SVC(), param_grid=param_grid, scoring='accuracy', cv=3)
```

```python
grid_search2.fit(x_train,y_train)
```

```python
grid_search2.best_params_
```
